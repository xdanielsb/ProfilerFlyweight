```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

## Memory plots

```python
csvFile = open('exports/mem-with-fw')
file = csv.reader(csvFile)
memory_FW = list(file)

csvFile = open('exports/mem-without-fw')
file = csv.reader(csvFile)
memory_without_FW = list(file)
```

```python
memory_FW = np.array(list(map(float,memory_FW[0])))
memory_without_FW = np.array(list(map(float,memory_without_FW[0])))
```

```python
plt.plot(range(1,len(memory_FW)+1),memory_FW/1000,label='with Flyweight')
plt.plot(range(1,len(memory_without_FW)+1),memory_without_FW/1000, label='without Flyweight')
plt.legend()
plt.ylabel('Consumed memory (kbytes)')
plt.xlabel('Number of soldiers')
plt.title("Memory consumption vs Number of Soldiers")
#plt.show()
plt.savefig("exports/mem_soldiers.png")
```

```python
csvFile = open('exports/mem-vary-color-length-with-fw')
file = csv.reader(csvFile)
memory_FW = list(file)

csvFile = open('exports/mem-vary-color-length-without-fw')
file = csv.reader(csvFile)
memory_without_FW = list(file)
```

```python
n_soldiers = [1,10,100,1000,1e6,1e7,1e8]
mem_fw = [152,368,2528,24128,24e6,240e6,2.4e9]
mem_not_fw = [136,1360,13600,136000,136e6,1360e6,13.6e9]
t_fw = [0.07,0.064,0.074,0.12,37.6,363,3525]
t_n_fw = [0.076,0.07,0.07,0.25,163.6,2053,16106]
```

```python
plt.plot(n_soldiers,mem_fw,label='with Flyweight')
plt.plot(n_soldiers,mem_not_fw, label='without Flyweight')
plt.legend()
plt.ylabel('Consumed memory (bytes)')
plt.xlabel('Number of soldiers')
plt.title("Memory consumption vs Number of Soldiers")
#plt.show()
plt.savefig("exports/mem_soldiers_v2.png")
```

```python
plt.plot(n_soldiers,t_fw,label='with Flyweight')
plt.plot(n_soldiers,t_n_fw, label='without Flyweight')
plt.legend()
plt.ylabel('Execution time (ms)')
plt.xlabel('Number of soldiers')
plt.title("Execution time vs Number of Soldiers")
#plt.show()
plt.savefig("exports/t_soldiers_v2.png")
```
